```python
import requests
import re
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

sitemap_url = 'https://metrkwadrat.net/post-sitemap.xml'

file_name = sitemap_url.lstrip('http://www.').rstrip('.net/post-sitemap.xml')

resoults_sitemap = requests.get(sitemap_url)

soup_sitemap = BeautifulSoup(resoults_sitemap.text, "lxml")
#print(soup_sitemap.prettify)

posts_html = []
titles = []
articles = []
n=0

links_html = soup_sitemap.find_all('loc')

for link in links_html:
    temp_link=link.text
    posts_html.append(temp_link)
    #print(temp_link)

# posts_html.remove('http://www.portalseo.pl/mapa-strony')
# posts_html.remove('http://www.portalseo.pl/mapa-strony/sad')
# posts_html.remove('http://www.portalseo.pl/')

# for post in posts_html[0:310]:
#     n+=1
#     resoult_post = requests.get(post)
#     soup_post = BeautifulSoup(resoult_post.text, 'lxml')
#     title_div = soup_post.find('h1', class_='post-title lg')
#     article_div = soup_post.find('div', class_='entry-content clearfix')
#     title = title_div.text
#     article_parts_txt= []
#     article_parts = article_div.find_all('p', style='text-align: justify;')
    
#     for parts in article_parts:
#         article_parts_txt.append(parts.text)
    
# #     article_parts_txt.remove(article_parts_txt[len(article_parts_txt)-1])
# #     article_parts_txt.remove(article_parts_txt[len(article_parts_txt)-1])
#     article_str = ' '.join(article_parts_txt)
# #     print(article_parts_txt)
# #     print(article_str)
        
#     if len(article_str) >= 1900 and len(article_str) <= 3100:
#         titles.append(title)
#         articles.append(article_str)
#         print(n,title, '\033[34m',len(article_str),'\033[32m','added','\033[0m')
    
#     else:
#         print(n,title, '\033[34m',len(article_str),'\033[31m','rejected','\033[0m')

        
for post in posts_html[359:360]:
    n+=1
    resoult_post = requests.get(post)
    soup_post = BeautifulSoup(resoult_post.text, 'lxml')
    title_div = soup_post.find('h1', class_='post-title lg')
    article_div = soup_post.find('div', class_='entry-content clearfix')
    title = title_div.text
    article_all = article_div.text
    
    #usówanie podpisów obrazków
    txt_to_rmv = []
    images_txt = article_div.find_all('figure')
    for txt in images_txt:
        txt_to_rmv.append(txt.text)
        print('   ',txt.text)
    for i in txt_to_rmv:
        article_all = article_all.replace(i,'')
        
    
    del_num = article_all.find('Tekst zewnętrzny')
    article_str = article_all[:del_num]
    print(article_str)
        
    if len(article_str) >= 1900 and len(article_str) <= 3100:
        titles.append(title)
        articles.append(article_str)
        print(n,title, '\033[34m',len(article_str),'\033[32m','added','\033[0m')
    
    else:
        print(n,title, '\033[34m',len(article_str),'\033[31m','rejected','\033[0m')
        
print('--------------------------------')
print('posts on site:','\033[34m',n,'\033[0m')
print('valuable articles:','\033[34m',len(articles),'\033[0m')
print('valuable titles:','\033[34m',len(titles),'\033[0m')

seo_articles = pd.DataFrame({
    'title': titles,
    'article': articles
})

print('--------------------------------')

if len(articles) > 0:
    seo_articles.to_csv(file_name+'_lng.csv')
    print('csv file','\033[32m','compiled','\033[0m')
    seo_articles.to_json(file_name+'_lng.json')
    print('json file','\033[32m','compiled','\033[0m')
else:
    print('\033[31m','0 articles to compile','\033[0m')
```
