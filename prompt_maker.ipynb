```python
#CORTICAL EXTRACT KEYWORDS

import os
import os.path
import requests
from time import sleep
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd

keywords_list = []
articles_en = []


path_to_file = input('EN file name:')
file_name = path_to_file.rstrip('_lng_en.xlsx')
file_en = pd.read_excel(path_to_file) 
articles_en = pd.DataFrame(file_en, columns=[' article']).values.tolist()
#for a in articles_en:
#    print('    ',a)


driver = webdriver.Chrome('/home/misiur/selenium_drivers/chromedriver')
driver.get('https://www.cortical.io/freetools/extract-keywords/')
driver.implicitly_wait(10)
driver.maximize_window()
reject_cookies = driver.find_element(By.ID, 'wt-cli-reject-btn')
reject_cookies.click()
text_input = driver.find_element(By.ID, 'text-input')

for article in articles_en:
    keywords_list_tmp = []
    keywords_str = ''
    text_input.send_keys(article)
    submit = driver.find_element(By.ID, 'submit-text')
    driver.implicitly_wait(60)
    submit.click()
    sleep(5)
    driver.implicitly_wait(60)
    keywords = driver.find_elements(By.CLASS_NAME, 'demo-keyword')
    for key in keywords:
        keywords_list_tmp.append(key.text)
    keywords_str = ', '.join(keywords_list_tmp)
    keywords_list.append(keywords_str)
    #print(keywords_str)
    text_input.clear()
    keywords.clear()

    
keys_en = pd.DataFrame({
    'key': keywords_list,
    })
    
sleep(1)
driver.quit()

a=0
for k in keywords_list:
    a+=1
    print(a,k)

save_path=('/home/misiur/scripts/scrapers/')
keys_en.to_excel(file_name+'keys_en.xlsx')

print('xlsx file','\033[32m','compiled','\033[0m')
```

```python
#json prompts
import json
import os
import requests
from time import sleep
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd

prompts = []
articles_str = []
articles_xlsx = input('aritcles file:')
keys_xlsx = input('keys file:')
articles_file = pd.read_excel(articles_xlsx)
keys_file = pd.read_excel(keys_xlsx)
articles = pd.DataFrame(articles_file, columns=['article']).values.tolist()
titles = pd.DataFrame(articles_file, columns=['title']).values.tolist()
keys = pd.DataFrame(keys_file, columns=['klucz']).values.tolist()
name = articles_xlsx.rstrip('_lng.xmls')

for ni in articles:
    art_str = ''.join(ni)
    articles_str.append(art_str)

for title in titles:
    n = titles.index(title)
    title_str = ' '.join(title)
    key_str = ' '.join(keys[n])
    prompt ='napisz artykuł SEO o tytule: '+title_str+'\nużyj:'+key_str
    prompts.append(prompt)
    
prompts_data  = pd.DataFrame({
    'prompt': prompts,
    'completion': articles_str,
})


prompts_data.to_json(name+'_prompts.json')
print('json file','\033[32m','compiled','\033[0m')
prompts_data.to_csv(name+'_prompts.csv')
print('csv file','\033[32m','compiled','\033[0m')
```
